# Feature Importance with Random Forest

This is a modified notebook from the [FastAI Machine Learning Course Lesson 2 notebook](https://github.com/fastai/fastai/blob/master/courses/ml1/lesson2-rf_interpretation.ipynb) and using the data from [Rossmann Store Sales](https://www.kaggle.com/c/rossmann-store-sales), an old Kaggle competition. The data was cleaned and pre-processed as per the [FastAI Deep Learning Course Lesson 3](https://github.com/fastai/fastai/blob/master/courses/dl1/lesson3-rossman.ipynb).

```python
from pandas_summary import DataFrameSummary
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from IPython.display import HTML
from fastai.structured import *
from fastai.column_data import *
from sklearn import metrics
```

```python
path = '/home/lyk/Documents/datasets/kaggle/rossmann'
```

```python
df_raw = pd.read_feather(f'{path}/joined')
```

```python
cat_vars = ['Store', 'DayOfWeek', 'Year', 'Month', 'Day', 'StateHoliday', 'CompetitionMonthsOpen',
    'Promo2Weeks', 'StoreType', 'Assortment', 'PromoInterval', 'CompetitionOpenSinceYear', 'Promo2SinceYear',
    'State', 'Week', 'Events', 'Promo_fw', 'Promo_bw', 'StateHoliday_fw', 'StateHoliday_bw',
    'SchoolHoliday_fw', 'SchoolHoliday_bw']

contin_vars = ['CompetitionDistance', 'Max_TemperatureC', 'Mean_TemperatureC', 'Min_TemperatureC',
   'Max_Humidity', 'Mean_Humidity', 'Min_Humidity', 'Max_Wind_SpeedKm_h', 
   'Mean_Wind_SpeedKm_h', 'CloudCover', 'trend', 'trend_DE',
   'AfterStateHoliday', 'BeforeStateHoliday', 'Promo', 'SchoolHoliday']

n = len(df_raw); n
```

```python
dep = 'Sales'
df_raw = df_raw[cat_vars+contin_vars+[dep, 'Date']].copy()
```

```python
for v in cat_vars: df_raw[v] = df_raw[v].astype('category').cat.as_ordered()
```

```python
for v in contin_vars:
    df_raw[v] = df_raw[v].astype('float32')
```

```python
df_raw = df_raw.set_index("Date")
```

`proc_df` replaces categories with their numeric codes, replaces missing values with median, splits the dependent variable and returns a dictionary of columns with missing values

```python
df, y, nas = proc_df(df_raw, 'Sales')
```

```python
m = RandomForestRegressor(n_jobs=-1)
%time m.fit(df,y)
m.score(df,y)
```

```python
def split_vals(a,n): return a[:n].copy(), a[n:].copy()

n_valid = int(len(df)*0.1)
n_train = len(df)-n_valid
X_train, X_valid = split_vals(df, n_train)
y_train, y_valid = split_vals(y, n_train)
raw_train, raw_valid = split_vals(df_raw, n_train)

X_train.shape, y_train.shape, X_valid.shape
```

Root Mean Square Percentage Error (RMSPE) is the metric used to evaluate the model in the Rossman Store Sales Kaggle competition.

```python
def exp_rmspe(y_pred, targ):
    pct_var = (targ - y_pred)/targ
    return math.sqrt((pct_var**2).mean())

def print_score(m):
    res = [exp_rmspe(m.predict(X_train), y_train), exp_rmspe(m.predict(X_valid), y_valid),
                m.score(X_train, y_train), m.score(X_valid, y_valid)]
    if hasattr(m, 'oob_score_'): 
        res.append(m.oob_score_)
    else: 
        res.append('NA')
    print(f'exp_rmspe of train:{res[0]},\nexp_rmspe of valid:{res[1]}, \nr2 of train:{res[2]}, \nr2 of valid:{res[3]}, \noob score:{res[4]}')
```

```python
%%time
m=RandomForestRegressor(n_jobs=-1, oob_score=True)
m.fit(X_train, y_train)
print_score(m)
```

```python
%%time
m=RandomForestRegressor(n_estimators=50, min_samples_leaf=3, max_features=0.5, n_jobs=-1, oob_score=True)
m.fit(X_train, y_train)
print_score(m)
```

Setting a leaf limit and sampling a subset of the features each time will make the model more generalizable. Couple this with increasing the number of estimators (trees), we were able to improve the model's score.

```python
fi = rf_feat_importance(m, df); fi[:10] #returns a sorted dataframe using the feature_importances_ method
```

```python
fi.plot('cols', 'imp', figsize=(10,6), legend=False);
```

```python
def plot_fi(fi): return fi.plot('cols', 'imp', 'barh', figsize=(12,7), legend=False)
```

```python
plot_fi(fi[:30]);
```

As shown above, there are some features such as temperature and humidity that are relatively lower score. We can remove them to simplify our model and reduce collinearity.

```python
to_keep = fi[fi.imp>0.005].cols; len(to_keep)
```

```python
df_keep = df[to_keep].copy()
X_train, X_valid = split_vals(df_keep, n_train)
```

```python
m = RandomForestRegressor(n_estimators=50, min_samples_leaf=3, max_features=0.5,n_jobs=-1, oob_score=True)
m.fit(X_train, y_train)
print_score(m)
```

```python
fi = rf_feat_importance(m, df_keep)
plot_fi(fi);
```

Removing some of the less important features reduces collinearity of the other features. In the above chart, some features have become more and less important when compared to the prior chart. For example, PromoInterval has become less important relative to other features like Day and Week.

### Hierarchical clustering

Hierarchical clustering helps us to see whether there are any features that are closely related and that we can remove.

```python
from scipy.cluster import hierarchy as hc
```

```python
corr = np.round(scipy.stats.spearmanr(df_keep).correlation, 4)
corr_condensed = hc.distance.squareform(1-corr)
z = hc.linkage(corr_condensed, method='average')
fig = plt.figure(figsize=(16,10))
dendrogram = hc.dendrogram(z, labels=df_keep.columns, orientation='left', leaf_font_size=16)
plt.show()
```

From the above chart, it would seem that the features 'Month' and 'Week', 'PromoInterval' and 'Promo2SinceYear' arer closely related.

```python
def get_oob(df):
    m = RandomForestRegressor(n_estimators=30, min_samples_leaf=5, max_features=0.6, n_jobs=-1, oob_score=True)
    x, _ = split_vals(df, n_train)
    m.fit(x, y_train)
    return m.oob_score_
```

```python
get_oob(df_keep)
```

```python
for c in ('Month', 'Week', 'PromoInterval', 'Promo2SinceYear'):
    print(c, get_oob(df_keep.drop(c, axis=1)))
```

```python
to_drop = ['Month']
get_oob(df_keep.drop(to_drop, axis=1))
```

```python
df_keep.drop(to_drop, axis=1, inplace=True)
X_train, X_valid = split_vals(df_keep, n_train)
```
